* найдите все такие столбцы (метод .isna() + подвыборка с условием)
* определите их названия (поймите как выцепить названия и сохраните их в массив, т.е. по цепочке примените методы .index, .tolist() или .values)
* "дропните их" (используйте метод .drop(имена столбцов))
* результат сохраните в новой табличке (например, data_1), исходную менять не нужно

```python
# находим столбцы с пропусками более 200
columns = data.isna().sum()[data.isna().sum() > 200].index.tolist()
columns
```

```python
# дропаем столбцы
data_1 = data.drop(columns, axis=1)
```

```python
# проверяем кол-во пропусков по столбцам в новой табличке
data_1.isna().sum()
```

### Задача 6

Создайте новый столбец TotalCost со значениями Quantity * UnitPrice. Сделайте resampling данных в еженедельные со значениями суммы по столбцам TotalCost.

```python
# создаем новый столбец
data['TotalCost'] = data.Quantity * data.UnitPrice
# делаем ресемплинг
data.TotalCost.resample('1W').sum()
```

### Задача 7

Посчитайте скользящее недельное среднее суммы ежедневных заказов. Для этого:
* Сделайте resampling в ежедневные данные
* Посчитайте скользящее среднее за 7 дней по столбцу TotalCost

```python
# делаем ресемплинг и считаем скользящее среднее по цепочке
data.TotalCost.resample('1D').sum().rolling('7D', min_periods=7).mean()
```

### Задача 8: GroupBy
